```python
import geopandas as gpd
import pandas as pd
from pathlib import Path

# --- Dossier contenant les trois sous-répertoires ---
BASE = Path(r"C:\Users\admin\Pictures\ISEP3\WDPA_WDOECM_Nov2025_Public_TCD_shp")

# --- Chemins des shapefiles de polygones ---
poly_paths = [
    BASE / "WDPA_WDOECM_Nov2025_Public_TCD_shp_0" / "WDPA_poly_0.shp",
    BASE / "WDPA_WDOECM_Nov2025_Public_TCD_shp_1" / "WDPA_poly_1.shp",
    BASE / "WDPA_WDOECM_Nov2025_Public_TCD_shp_2" / "WDPA_poly_2.shp",
]

# --- Lecture des trois jeux de données ---
gdf_polys_list = [gpd.read_file(p) for p in poly_paths]

# Contrôle du système de coordonnées
for i, gdf in enumerate(gdf_polys_list):
    print(f"CRS partie {i+1} :", gdf.crs)

# --- Fusion des enregistrements ---
gdf_polys_merged = gpd.GeoDataFrame(
    pd.concat(gdf_polys_list, ignore_index=True),
    crs=gdf_polys_list[0].crs
)

print("Nombre total de polygones :", len(gdf_polys_merged))

# --- Export du shapefile fusionné ---
out_poly = BASE / "WDPA_TCD_polygons_merged.shp"
gdf_polys_merged.to_file(out_poly)

print("Fichier exporté :", out_poly)
```
